```python
import numpy as np
import os
import sys
import sklearn.feature_extraction as skl_feat_extrac
import pandas as pd

import glob
from skimage import io
import matplotlib.pyplot as plt
import shutil
```

# Saving Unique Intensity Values Number for Patches in an Excel Files

```python
#--------------------------------------------------------------------
patches_path = 'E:\\Python files\\UdG\\Thesis\\Datasets\\OPTIMAM\\patches\\**\\*.png'

# find images names
patches_names = glob.glob(patches_path + '\\**\\*.png', recursive=True)

intensity_dist = {'BatchID':[], 'ClientID':[], 'StudyID':[],'ImageID': [], 'PatchID':[], 'NumUniqueValues':[]}

print(intensity_dist.keys(), intensity_dist.values())
print(os.path.exists(patches_path))

print(len(patches_names))

# loop over all images
for idx, name in enumerate( patches_names):
    
    # find how many different intensity values are there for the current image
    dist = len(np.unique(io.imread(name)))
    properties = name.rsplit('\\')
    
    # read the properties from the absolute image name
    intensity_dist['BatchID'].append(properties[7])
    intensity_dist['ClientID'].append(properties[8])
    
    intensity_dist['StudyID'].append(properties[9])
    intensity_dist['ImageID'].append(properties[10])
    
    intensity_dist['PatchID'].append(properties[11])
    
    intensity_dist['NumUniqueValues'].append(dist)
    print(str(idx), 'done')
    

print('-'*8 + 'DONE' + '-'*8)

# save these properties as an excel file
intensity_dataframe = pd.DataFrame(intensity_dist)
intensity_dataframe.head()
pd.DataFrame.to_excel(intensity_dataframe, patches_path + '\\intensity distributions.xlsx')
```

# Showing Patches With Small Number of Different Values
# Deleting them

```python
patches_path = 'E:\\Python files\\UdG\\Thesis\\Datasets\\OPTIMAM\\patches'

# read the excel file
data_frame = pd.read_excel(patches_path + '\\intensity distributions.xlsx')
threshold = 30

# filter the images using number of unique values column
img_names_frame = data_frame[data_frame['NumUniqueValues'] < threshold]  #.isin(range(threshold, threshold+10))
print(img_names_frame.shape)

batch_sample = img_names_frame.sample(frac=1)#n= np.minimum(100, img_names_frame.shape[0])

# for all the rows in the filtered excel file
for row in range(batch_sample.shape[0]): #
        batch_num = img_names_frame.iloc[row]['BatchID']
        patien_id = img_names_frame.iloc[row]['ClientID']
    
        study_id = img_names_frame.iloc[row]['StudyID']
        image_id = img_names_frame.iloc[row]['ImageID']
        
        patch_id = img_names_frame.iloc[row]['PatchID']
        unique_val = img_names_frame.iloc[row]['NumUniqueValues']
        
        #construct the file name from excel records
        file_name = os.path.join(patches_path, str(batch_num), str(patien_id), str(study_id), str(image_id), patch_id)
        print(file_name, '\n', row)
        
        if os.path.exists(file_name):
            
            # read the image and show it
            plt.figure()
            img = io.imread(file_name)

            plt.imshow(img, cmap='gray')
            plt.title(str(patch_id) + ' '+ str(unique_val) + ' '+ str(row))
        else:
            print('deleted successfully', row)        
```

# Extracting Files from Their Nests

```python
"""
This file is to copy specific lesion type patches from a source folder to a destination folder
"""
dst_path = 'E:\\Python files\\UdG\\Thesis\\Datasets\\OPTIMAM\\classifiedInput\\Cal\\all'
src_path = 'E:\\Python files\\UdG\\Thesis\\Datasets\\OPTIMAM\\input\\all\\1Les\\*.png'

# fetch images names
file_names = glob.glob(src_path, recursive=True )
types = np.array([])

#the type of lesion desired 
specific_type = 'Cal'

count = 0
file_names = np.array(file_names)
np.random.shuffle(file_names)

#for all files
for idx, name in  enumerate(file_names):
    
    file_name = os.path.split(name)[-1] #get the last part after the last '/'
    les_type = str.rsplit(file_name, sep='_')[-2] #get the part related to lesion type
    
    types = np.append(types, les_type)
    
    # if the type of the read file matches the desired one, copy it
    if les_type == specific_type:
        count+=1
        shutil.copy(name, dst_path)
        print('finished ','file name', file_name, 'type', les_type, idx, count)
        
    
print('types shape', np.shape(types))
unique_types, counts = np.unique(types, return_counts=True)
    
print('unique types', unique_types, ' counts', counts)
```
